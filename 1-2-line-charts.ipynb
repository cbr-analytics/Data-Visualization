## Learn python to create a nice professional line charts, using a real-world example.

We use daily streams from the music streaming service 'SPOTIFY'.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_spotify = pd.read_csv('data/spotify.csv')
df_spotify.head()
```

    - The data shows, songs in columns versus date-wise view count on the on-line streaming platform.
Let's make 'Date' column as index.

```python
df_spotify = pd.read_csv('data/spotify.csv',
                        index_col='Date', parse_dates = True)
df_spotify.head()
```

```python
#Check tail of the data
df_spotify.tail()
```

## Line plot

```python
sns.lineplot(data= df_spotify)
```

## Format the line plot

```python
#Set figure size
plt.figure(figsize=(16,8))

# Add title
plt.title("Daily Global Streams of Songs")

# line chart
sns.lineplot(data=df_spotify)
```
